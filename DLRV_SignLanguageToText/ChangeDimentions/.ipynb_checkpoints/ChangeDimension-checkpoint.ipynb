```python
import numpy as np
import pandas as pd
import cv2
import csv

def resize_to_28x28(img):
    "Taken from:https://stackoverflow.com/questions/66711654/converting-image-into-mnist-format"
    img_h, img_w = img.shape
    dim_size_max = max(img.shape)

    if dim_size_max == img_w:
        im_h = (26 * img_h) // img_w
        if im_h <= 0 or img_w <= 0:
            print("Invalid Image Dimention: ", im_h, img_w, img_h)
        tmp_img = cv2.resize(img, (26,im_h),0,0,cv2.INTER_NEAREST)
    else:
        im_w = (26 * img_w) // img_h
        if im_w <= 0 or img_h <= 0:
            print("Invalid Image Dimention: ", im_w, img_w, img_h)
        tmp_img = cv2.resize(img, (im_w, 26),0,0,cv2.INTER_NEAREST)

    out_img = np.zeros((28, 28), dtype=np.ubyte)

    nb_h, nb_w = out_img.shape
    na_h, na_w = tmp_img.shape
    y_min = (nb_w) // 2 - (na_w // 2)
    y_max = y_min + na_w
    x_min = (nb_h) // 2 - (na_h // 2)
    x_max = x_min + na_h

    out_img[x_min:x_max, y_min:y_max] = tmp_img

    return out_img

"""
def run_inference(img):
    tsr_img = resize_to_28x28(img)
    input_data = np.copy(tsr_img).reshape(1,28,28,1)

    if floating_model:
        input_data = (np.float32(input_data) - input_mean) / input_std

    # Instantiate tensorflow interpreter and run inference on input_data
"""
def change_dimensions(read_path,write_path,samples,labels,filename):
    for letter in labels:
        for i in range(0,samples):
            importedimage = cv2.imread(read_path+'/'+letter+"/"+letter+"_"+str(i)+".jpg")
            gray = cv2.cvtColor(importedimage, cv2.COLOR_BGR2GRAY)
            new_image = resize_to_28x28(gray)
            new_image_flatten = new_image.flatten()
            new_image_with_label = np.insert(new_image_flatten,0,int(labels.index(letter)))
            with open(write_path+filename+".csv",'a') as f:
                writer = csv.writer(f)
                writer.writerow(new_image_with_label)

        print(letter+" is converted")
        
def save_train_test_file(path,filename):
    head = ['label']
    for i in range(1,785):
        head.append('pixel'+str(i))
    df = pd.read_csv(path+filename+".csv")
    ds = df.sample(frac=1,axis=0)
    ds.to_csv(path+filename+'_shuffel.csv',header=head,mode='w',index=False)
    
    msk = np.random.rand(len(ds)) <= 0.7

    train = ds[msk]
    test = ds[~msk]

    train.to_csv(path+'collected_train.csv',header=head,mode='w',index=False)
    test.to_csv(path+'collected_test.csv',header=head,mode='w',index=False)
```

```python
!mkdir ../Data
read_path = '../CollectedImages/'
write_path = '../Data/'
samples = 50
labels = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
filename = 'convertedImages'
change_dimensions(read_path=read_path,write_path=write_path,samples=samples,labels=labels,filename=filename)
save_train_test_file(path=write_path,filename=filename)
```
